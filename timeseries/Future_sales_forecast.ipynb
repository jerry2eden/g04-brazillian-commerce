```python
#importing necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from fbprophet import Prophet

%matplotlib inline
```

```python
#loading the necessary datasets into pandas dataframes

orders_df = pd.read_csv('../data/olist_orders_dataset.csv')
order_items_df = pd.read_csv('../data/olist_order_items_dataset.csv')
products_df = pd.read_csv('../data/olist_products_dataset.csv')
```

```python
#converting columns containing dates to datetime dtype

for col in ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date']:
    orders_df[col] = pd.to_datetime(orders_df[col])
```

```python
#merging our dataframes to enable us get the information we need

customers_order_items_df = pd.merge(orders_df, order_items_df, on='order_id', how='left')
customers_order_items_df = pd.merge(customers_order_items_df, products_df, on='product_id', how='left')
customers_order_items_df
```

```python
time_series_df = customers_order_items_df.set_index('order_purchase_timestamp')
```

```python
time_series_df = time_series_df[['order_id', 'price']]
time_series_df
```

```python
#converting our dataframe to show how much was earned daily

daily_revenue_df = time_series_df.resample('D').sum()
daily_revenue_df.reset_index(inplace=True)
daily_revenue_df
```

```python
daily_revenue_df.price.sum()    #should be equal to the total revenue we had before resampling (check EDA.ipynb)
```

```python
#converting our dataframe to see how many orders we received daily (remember, this is order items, as some orders contain multiple items)

daily_orders_df = time_series_df.resample('D').count()
daily_orders_df.drop('price', axis=1, inplace=True)
daily_orders_df.reset_index(inplace=True)
daily_orders_df
```

```python
daily_orders_df.order_id.sum()     #should be equal to the total number of order items before resampling (check EDA.ipynb)
```

We would be using Facebook-Prophet for our time series forecasting.

Facebook Prophet requires that you name your date column 'ds' and your target column 'y'

Our forecast is univariate as the available dataset would not permit us to work with multivariate time series.

```python
daily_revenue_df.columns = ['ds', 'y']
daily_revenue_df
```

```python
daily_orders_df.columns = ['ds', 'y']
daily_orders_df
```

As observed during EDA, transactions after a certain date do not have price information. Including data from these dates would affect our forecast negatively.

Also, I decided to use exactly 2 years worth of data to fit the model (this gives 730 days).

Factoring these considerations, I decided to drop data from dates after "4th September, 2018". This works out well for our considerations.

```python
daily_revenue_df[daily_revenue_df['ds']=='2018-09-04']
```

```python
#As we can see, orders after "4th September, 2018" in our dataset lack price information

daily_revenue_df[730:]
```

```python
daily_revenue_df.drop(index=np.arange(730, 774), inplace=True)
```

```python
daily_orders_df.drop(index=np.arange(730, 774), inplace=True)
```

```python
daily_orders_df
```

```python
revenue_model = Prophet(yearly_seasonality=True)
revenue_model.fit(daily_revenue_df)
```

```python
future = revenue_model.make_future_dataframe(periods= 365 , freq= 'D' )
revenue_forecast = revenue_model.predict(future)
revenue_forecast
```

```python
revenue_model.plot(revenue_forecast)
plt.show()
```

```python
revenue_model.plot_components(revenue_forecast)
```

```python
predicted_revenue_df = revenue_forecast[['ds', 'yhat']]
predicted_revenue_df
```

```python
predicted_revenue_df[predicted_revenue_df['ds']=='2018-09-04']
```

```python
predicted_revenue_df = predicted_revenue_df[730:]
predicted_revenue_df.columns = ['date', 'revenue']
```

```python
predicted_revenue_df.set_index('date', inplace=True)
predicted_revenue_df
```

```python
predicted_revenue_df.plot();
```

```python
daily_revenue_df.columns = ['date', 'revenue']
daily_revenue_df.set_index('date', inplace=True)
daily_revenue_df
```

```python
plt.rcParams["figure.figsize"] = (16,9)
fig, ax = plt.subplots()
ax.plot(daily_revenue_df)
ax.plot(predicted_revenue_df, label='predicted revenue')
leg = ax.legend(loc ="upper right");
```
